3. Importe de la data a VS Code.

```python
# 3. Importe de datos

# Leer el archivo Parquet
data = pd.read_parquet("C:/HENRY DATAFT-10/M4 - Python/PI-4/data_latinoamerica.parquet")

# Mostrar los primeros registros del DataFrame
print(data.head())
```

```python
# Validamos la escritura de cada uno de los paises con .unique() para crear la lista que nos servira para el filtrado.

data['country_name'].unique()
```

5. Filtrado por paises requeridos.

```python
# 5. Filtrado de paises.

# Filtramos nuestra la data por los paises requeridos con .isin()

paises_seleccionados = ['Colombia', 'Argentina', 'Chile', 'Mexico', 'Peru', 'Brazil']       #Declaramos la lista paises_seleccioandos
data_latinoamerica = data[data['country_name'].isin(paises_seleccionados)]

print(data_latinoamerica['date'])
```

Filtramos nuevamente el DataFrame utilizando la columna 'location_key', pero esta vez nos enfocamos únicamente en las instancias que corresponden a los países seleccionados (Argentina, Chile, Colombia, México, Perú y Brasil), excluyendo cualquier subregión o región específica de estos países. Para ello, nos aseguramos de que los valores en location_key coincidan exactamente con los códigos de país: AR (Argentina), CL (Chile), CO (Colombia), MX (México), PE (Perú) y BR (Brasil).

```python
data_latinoamerica = data_latinoamerica[data_latinoamerica['location_key'].isin(['AR','CL','CO','MX','PE','BR'])]
print(data_latinoamerica.shape)
```

6. Filtrado de los datos por las fechas iguales o mayores a 2021-01-01.

```python
# 6. Filtramos los datos en fechas mayores a 2021-01-01.

data_latinoamerica = data_latinoamerica[data_latinoamerica['date']>='2021-01-01']     #Aplicamos una mascara para el filtrado por fecha.

print(data_latinoamerica)      # Nuestro dataframe conservó el mismo nombre ya filtrado por fechas.
```

7. Manejo de nulos. Los reemplazamos con la media.

```python
# Sumamos la cantidad de valores nulos en cada columna.
nulos_por_columna = data_latinoamerica.isnull().sum()

# Mostramos la cantidad de nulos por columna
print(nulos_por_columna)
```

```python
#Seleccionamos las columnas con una cantidad de nulos mayor al 80% (3000) para validar si tenemos que eliminarlas.

nulos = data_latinoamerica.isnull().sum()[data_latinoamerica.isnull().sum()>3000]       

print(nulos)

# Vemos que ninguna supera el 80% de nulos entonces no debemos eliminar ninguna.
```

```python

# Eliminamos filas donde todos los valores son nulos
data_latinoamerica = data_latinoamerica.dropna(axis=0, how='all')

# Mostrar el DataFrame resultante
print(data_latinoamerica)
```

Probar con la moda o mediana.

```python
# Lista de columnas a rellenar
columnas_nulos = ['new_confirmed', 'new_deceased', 'cumulative_confirmed', 'cumulative_deceased', 'cumulative_vaccine_doses_administered',
                  'average_temperature_celsius', 'minimum_temperature_celsius', 'maximum_temperature_celsius', 'rainfall_mm',
                  'relative_humidity', 'new_recovered', 'cumulative_recovered']

# Filtramos solo las columnas que existen en el DataFrame
columnas_existentes = [col for col in columnas_nulos if col in data_latinoamerica.columns]

# Agrupamos por país y calculamos la media de cada columna por país
media_por_pais = data_latinoamerica.groupby('country_name')[columnas_existentes].transform(lambda x: x.mean(skipna=True))

# Rellenamos los valores nulos con la MEDIA correspondiente a cada país
data_latinoamerica[columnas_existentes] = data_latinoamerica[columnas_existentes].fillna(media_por_pais)

# Mostrar el DataFrame resultante
print(data_latinoamerica)
```

```python
# Llenar Nan con el valor de 0 en la columna new_recovered, cumulative_recovered, para hacer analisis de estos datos.
data_latinoamerica['new_recovered']=data_latinoamerica['new_recovered'].fillna(0)
data_latinoamerica['cumulative_recovered']=data_latinoamerica['cumulative_recovered'].fillna(0)

# Mostrar el DataFrame resultante
print(data_latinoamerica)
```

```python
# Aqui verificamos que cada columna con valores nulos haya sido rellenada.
print(data_latinoamerica.isnull().sum())
data_latinoamerica.shape
```

8. Guardado del dataframe tratado para usos futuros.


```python
#Renombramos nuestro dataframe y lo guardamos.
data_latinoamerica_final = data_latinoamerica
data_latinoamerica_final.to_csv('data_latinoamerica_final.csv', index=False)
```

1. Distribución de la población por grupos de edad

```python
# Definimos las columnas de población por edad
columnas_edad = ['population_age_00_09', 'population_age_10_19', 'population_age_20_29','population_age_30_39', 'population_age_40_49',
                 'population_age_50_59','population_age_60_69', 'population_age_70_79', 'population_age_80_and_older']

# Sumamos las poblaciones por grupo de edad para cada país
poblacion_por_edad = data_latinoamerica_final.groupby('country_name')[columnas_edad].sum()

# Normalizamos los datos para obtener proporciones
poblacion_por_edad = poblacion_por_edad.div(poblacion_por_edad.sum(axis=1), axis=0) * 100

plt.figure(figsize=(14, 8))
ax = poblacion_por_edad.plot(kind='bar', stacked=True, colormap='viridis', figsize=(14, 8))
plt.title('Distribución de la población por grupos de edad en Latinoamérica', fontsize=16)
plt.xlabel('Países', fontsize=14)
plt.ylabel('Proporción de la población (%)', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Grupos de edad', bbox_to_anchor=(1.05, 1), loc='upper left')

# Añadimos los porcentajes dentro de cada barra
for i, patch in enumerate(ax.patches):
    # Obtenemos las coordenadas y el porcentaje correspondiente
    height = patch.get_height()
    if height > 0:
        ax.text(patch.get_x() + patch.get_width() / 2, patch.get_y() + height / 2,
                '{:.1f}%'.format(height), ha='center', va='center', fontsize=11, color='white', rotation=0)

plt.show()
```

9. Correlación de indicadores demográficos y de salud

```python
# Seleccionamos las variables relevantes (igual que antes)
variables_demograficas = [
    'gdp_per_capita_usd',        # PIB per cápita
    'human_development_index',   # IDH
    'population_density'         # Densidad poblacional
]

variables_salud = [
    'diabetes_prevalence',           # Prevalencia de diabetes
    'smoking_prevalence',            # Prevalencia de tabaquismo
    'life_expectancy',               # Esperanza de vida
    'comorbidity_mortality_rate',    # Mortalidad por comorbilidades
    'physicians_per_1000',           # Médicos por 1000 habitantes
    'nurses_per_1000'                # Enfermeros por 1000 habitantes
]

# Calculamos la matriz de correlación
matriz_correlacion = data_latinoamerica_final[variables_demograficas + variables_salud].corr()

# Filtramos las correlaciones cruzadas: SALUD (filas) vs DEMOGRÁFICOS (columnas)
matriz_cruzada = matriz_correlacion.loc[variables_salud, variables_demograficas]  # Cambiamos el orden aquí

# Configuramos el gráfico
plt.figure(figsize=(10, 8))
sns.heatmap(
    matriz_cruzada,
    annot=True,
    cmap='flare',
    vmin=-1, 
    vmax=1,
    fmt='.2f',
    linewidths=0.5
)

# Personalizamos el gráfico
plt.title('Correlación entre indicadores de salud y demográficos', fontsize=16)
plt.xlabel('Indicadores demográficos', fontsize=12)  # Eje X ahora son variables demográficas
plt.ylabel('Indicadores de salud', fontsize=12)      # Eje Y ahora son variables de salud
plt.xticks(rotation=15, ha='right')
plt.yticks(rotation=0)               # Eje Y (salud) sin rotación

plt.tight_layout()  # Ajusta el espacio para evitar cortes en las etiquetas
plt.show()
```

10. Comparación de la Prevalencia de Diabetes y la Tasa de Mortalidad

```python
# Preparamos los datos (agrupamos por país)
datos_agrupados = data_latinoamerica_final.groupby('country_name').agg({
    'diabetes_prevalence': 'mean',
    'comorbidity_mortality_rate': 'mean'
}).reset_index()

# Ordenamos los países por prevalencia de diabetes (opcional, mejora la legibilidad)
datos_agrupados = datos_agrupados.sort_values('diabetes_prevalence', ascending=False)

# Configuramos el gráfico
plt.figure(figsize=(10, 6))

# Posiciones de las barras
paises = datos_agrupados['country_name']
posiciones = np.arange(len(paises))
ancho_barras = 0.35  # Ancho de cada barra

# Barras para Diabetes
barras_diabetes = plt.bar(
    posiciones - ancho_barras/2, 
    datos_agrupados['diabetes_prevalence'], 
    width=ancho_barras, 
    label='Prevalencia de Diabetes (%)',
    color='#FF6B6B'  # Rojo pastel
)

# Barras para Mortalidad
barras_mortalidad = plt.bar(
    posiciones + ancho_barras/2, 
    datos_agrupados['comorbidity_mortality_rate'], 
    width=ancho_barras, 
    label='Tasa de Mortalidad por Comorbilidades',
    color='#4B2362'  # Verde agua
)

# Personalización
plt.title('Comparación de Diabetes y Mortalidad por Comorbilidades en Países de Latinoamérica', fontsize=16, pad=20)
plt.xlabel('Países', fontsize=12)
plt.ylabel('Valor (%)', fontsize=12)
plt.xticks(posiciones, paises, rotation=45, ha='right')  # Rotación para legibilidad


# Añadimos los valores encima de las barras
for barra in barras_diabetes:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 0.2, 
             f'{altura:.1f}', ha='center', va='bottom', fontsize=9)

for barra in barras_mortalidad:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 0.2, 
             f'{altura:.1f}', ha='center', va='bottom', fontsize=9)

# Leyenda y ajustes finales
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()  # Mejora el espaciado
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convertir la columna 'date' a tipo datetime
data_latinoamerica_final['date'] = pd.to_datetime(data_latinoamerica_final['date'])

# Filtrar datos para 2021 y 2022
datos_2021 = data_latinoamerica_final[data_latinoamerica_final['date'].dt.year == 2021]
datos_2022 = data_latinoamerica_final[data_latinoamerica_final['date'].dt.year == 2022]

# Agrupar por mes y calcular la suma de casos activos y recuperados
# Casos activos = nuevos confirmados - nuevos recuperados - nuevos fallecidos
datos_2021['casos_activos'] = datos_2021['new_confirmed'] - datos_2021['new_recovered'] - datos_2021['new_deceased']
datos_2022['casos_activos'] = datos_2022['new_confirmed'] - datos_2022['new_recovered'] - datos_2022['new_deceased']

# Agrupar por mes
datos_2021_mes = datos_2021.groupby(datos_2021['date'].dt.month).agg({
    'casos_activos': 'sum',
    'new_recovered': 'sum'}).reset_index()

datos_2022_mes = datos_2022.groupby(datos_2022['date'].dt.month).agg({
    'casos_activos': 'sum',
    'new_recovered': 'sum'
}).reset_index()

# Renombrar la columna de meses para que sea más clara
datos_2021_mes.rename(columns={'date': 'mes'}, inplace=True)
datos_2022_mes.rename(columns={'date': 'mes'}, inplace=True)

# Crear los gráficos
plt.figure(figsize=(14, 6))

# Gráfico para 2021
plt.subplot(1, 2, 1)
sns.lineplot(data=datos_2021_mes, x='mes', y='casos_activos', label='Casos Activos', color='red')
sns.lineplot(data=datos_2021_mes, x='mes', y='new_recovered', label='Casos Recuperados', color='green')
plt.title('Evolución de Casos Activos vs. Recuperados (2021)')
plt.xlabel('Mes')
plt.ylabel('Cantidad de casos')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend()

# Gráfico para 2022
plt.subplot(1, 2, 2)
sns.lineplot(data=datos_2022_mes, x='mes', y='casos_activos', label='Casos Activos', color='red')
sns.lineplot(data=datos_2022_mes, x='mes', y='new_recovered', label='Casos Recuperados', color='green')
plt.title('Evolución de Casos Activos vs. Recuperados (2022)')
plt.xlabel('Mes')
plt.ylabel('Cantidad de casos')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend()

plt.tight_layout()
plt.show()
```

Aqui vemos una tendencia mayor en la cantidad de recuperados sobre la cantidad de casos activos, sin embargo podemos analizar que en el mes Noviembre del 2021 hasta finales de Enero del 2022 empezaron a subir los casos activos por sobre los recuperados. Mi apresiacion personal es que esto pudo haber ocurrido debido a las festividades de fin de año.

```python
# Asegurarnos de que la columna 'date' esté en formato datetime
data_latinoamerica_final['date'] = pd.to_datetime(data_latinoamerica_final['date'])

# Agrupar por fecha y calcular la media mensual de 'new_confirmed' y 'average_temperature_celsius'
data_mensual = data_latinoamerica_final.resample('ME', on='date').mean(numeric_only=True)

# Crear el gráfico
plt.figure(figsize=(20, 7))
ax1 = plt.gca()  # Primer eje y (nuevos casos confirmados)
ax2 = ax1.twinx()  # Segundo eje y (temperatura promedio)

# Graficar nuevos casos confirmados (línea roja punteada)
ax1.plot(
    data_mensual.index, 
    data_mensual['new_confirmed'], 
    color='red', 
    label='Nuevos casos confirmados',
    linestyle=':',  # Línea punteada
    linewidth=2  # Hacer la línea un poco más gruesa
)

# Graficar temperatura promedio (línea azul continua)
ax2.plot(
    data_mensual.index, 
    data_mensual['average_temperature_celsius'], 
    color='blue', 
    label='Temperatura promedio (°C)', 
    linestyle='-',  # Línea continua
    linewidth=2  # Hacer la línea un poco más gruesa
)

# Etiquetas y título con fuentes más grandes
ax1.set_xlabel('Fecha', fontsize=14)
ax1.set_ylabel('Nuevos casos confirmados', color='red', fontsize=14)
ax2.set_ylabel('Temperatura promedio (°C)', color='blue', fontsize=14)
plt.title('Nuevos casos de COVID-19 y temperatura promedio con el tiempo', fontsize=16, pad=20)

# Leyendas con fuentes más grandes
ax1.legend(loc='upper left', fontsize=12)
ax2.legend(loc='upper right', fontsize=12)

# Ajustar el tamaño de las etiquetas de los ejes
ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Hemos notado una tendencia de incremento en la temperatura promedio, mientras que los casos confirmados de COVID-19 han disminuido. Sin embargo, entre enero y febrero del 2022 se presentó un repunte que descendió rápidamente hacia marzo. Este fenómeno podría atribuirse al número de vacunas aplicadas o al incremento en la cantidad de personas vacunadas.
